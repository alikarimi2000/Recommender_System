```python
def get_top_similar_movi(movi_title, movie_similarity_df):
   
    sorted_user_corr = movie_similarity_df.loc[movi_title].sort_values(ascending=False)
    top_similar_users = sorted_user_corr.iloc[1:30].index

    result_dict = {'title': [], 'Similar title': [], 'Correlation': []}
    

    for similar_user in top_similar_users:
        correlation = movie_similarity_df.loc[movi_title, similar_user]
        result_dict['title'].append(movi_title)
        result_dict['Similar title'].append(similar_user)
        result_dict['Correlation'].append(correlation)

    result_df = pd.DataFrame(result_dict)
    
    return result_df
```

```python
similar_title=get_top_similar_movi('Star Wars (1977)',movie_similarity_df)
```

```python
similar_title.head(10)
```

```python
def prediction(similar_title,user_id,movi):
    total_sum=0
    total_coor=0
    dic={}
    for item in similar_title.values:
        title = item[1]
        similarity_score = item[2]

        if pd.notna(moviemat.loc[user_id,title]):
            total_sum += moviemat.loc[user_id,title] * similarity_score
            total_coor+=similarity_score

    if total_sum!=0:
        dic['title']=movi
        dic['Total_Sum']=total_sum/total_coor
    else:
        dic['title']=movi
        dic['Total_Sum']=total_sum

    return dic
```

```python
def calculate_corr(input_dataframe, target_column):
    # Calculate the correlation between the target column and other columns in the input dataframe.
    similar = input_dataframe.corrwith(input_dataframe[target_column])

    corr_dataframe = pd.DataFrame(similar, columns=['Correlation'])
    corr_dataframe.dropna(inplace=True)
    

    top_corr_values = corr_dataframe.sort_values('Correlation', ascending=False).reset_index()
    
    top_corr_values.insert(0, 'target_title', target_column)
    
    top_corr_values.rename(columns={'index': 'Similar title', 'Correlation': 'Similarity'}, inplace=True)
    
    return top_corr_values.head(30)


```
