```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
```

```python
folder_path = 'D:\\IASBS\\Data_mining\\project\\ml-100k\\'
```

```python
#Read data from the dataset file into a Pandas DataFrame with tab-separated values
columns = ['user_id', 'item_id', 'rating', 'timestamp']
#df = pd.read_csv(folder_path+'u.data', sep='\t', names=columns)
df = pd.read_csv(folder_path+'u1.base', sep='\t', names=columns)
df.head()
```

```python
movies_df = pd.read_csv(folder_path + 'u.item', sep='|', header=None, encoding='latin-1')
```

```python
movies_df 
```

```python
movies_df=movies_df.iloc[: , 0:2]
movies_df.columns=['item_id','title']
```

```python
df = pd.merge(df, movies_df, on='item_id')
df.head()
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat_notna=moviemat.fillna(0)
moviemat.head()
```

```python
user_ratings_12_angry_men = moviemat['12 Angry Men (1957)']
df_users_rated_12_angry_men = moviemat_notna[user_ratings_12_angry_men.notnull()]
user_ratings_12_angry_men.dropna(inplace=True)
df_users_rated_12_angry_men.head()

```

```python
columns = ['user_id', 'item_id', 'rating', 'timestamp']
df_test = pd.read_csv(folder_path+'ua.test', sep='\t', names=columns)
df_test.head()
```

```python
df_test = pd.merge(df_test, movies_df, on='item_id')
df_test['num_votes'] = df_test.groupby('title')['rating'].transform('count')
```

```python
moviemat_test = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat_test_notna=moviemat_test.fillna(0)
moviemat_test.head()
```

```python
target_user = moviemat_test_notna.iloc[[0]]
```

```python

user_knn = KNeighborsRegressor(metric='cosine', n_neighbors=10)
user_knn.fit(df_users_rated_12_angry_men, user_ratings_12_angry_men)
user_user_pred = user_knn.predict(target_user)
print(user_user_pred)

```
