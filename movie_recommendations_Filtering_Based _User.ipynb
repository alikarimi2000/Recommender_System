```python

movies_df = pd.read_csv(folder_path + 'u.item', sep='|', header=None, encoding='latin-1')

```

```python
#Selecting only the first two columns
movies_df=movies_df.iloc[: , 0:2]
movies_df.columns=['item_id','title']
```

```python
df = pd.merge(df, movies_df, on='item_id')
df.head()
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
moviemat.shape
```

```python
user_corr = moviemat.T.corr(min_periods=20)
print(user_corr)
```

```python
corr_user = pd.DataFrame(user_corr )
```

```python
corr_user.head(100)
```

```python

#Get the top similar users for a given user based on user correlation
def get_top_similar_users(user_id, user_corr_df,num_users=30):


    sorted_user_corr = user_corr_df.loc[user_id].sort_values(ascending=False)
    top_similar_users = sorted_user_corr.iloc[1:num_users].index

    result_dict = {'User': [], 'Similar User': [], 'Correlation': []}
    

    for similar_user in top_similar_users:
        correlation = user_corr_df.loc[user_id, similar_user]
        result_dict['User'].append(user_id)
        result_dict['Similar User'].append(similar_user)
        result_dict['Correlation'].append(correlation)

    result_df = pd.DataFrame(result_dict)
    
    return result_df
```

```python
user_id=1
similar_users=get_top_similar_users(user_id,corr_user)
```

```python
similar_users.head(100)
```

```python
dic={}
#  for Loop each movie in the moviemat columns
for movie in moviemat.columns:
    total_sum = 0
    total_coor=0
    # Loop through each item in the similar_users values
    for item in similar_users.values:
        user_id = item[1]
        similarity_score = item[2]
         # Check if the movie rating for the user is not NaN
        if pd.notna(moviemat.loc[user_id, movie]):
            total_sum += moviemat.loc[user_id, movie] * similarity_score
            total_coor+=similarity_score

        if total_sum!=0:
            dic[movie]=total_sum/total_coor
        else:
            dic[movie]=total_sum
dic = {'title': list(moviemat.columns), 'predict': list(dic.values())}
predict = pd.DataFrame(dic)
predict.head()

```

```python
def calculate_prediction_error(predictions_df, actual_ratings_df):


    merged_df = pd.merge(predictions_df, actual_ratings_df, on='title')

    merged_df['squared_error'] = (merged_df['predict'] - merged_df[merged_df.columns[2]])**2
    merged_df.head()

    mse = merged_df['squared_error'].mean()
    rmse = np.sqrt(mse)

    return mse, rmse


mse, rmse = calculate_prediction_error(predict , user)
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python

def calculate_squared_error(predictions_df, actual_ratings_df):

    merged_df = pd.merge(predictions_df, actual_ratings_df, on='title')
    merged_df['squared_error'] = (merged_df['predict'] - merged_df[merged_df.columns[2]]) ** 2

    return merged_df


error_df = calculate_squared_error(predict, user)

```
